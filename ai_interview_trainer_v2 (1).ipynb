```python
import json
import os
from datetime import datetime

MEMORY_FILE = "interview_memory.json"

if os.path.exists(MEMORY_FILE):
    with open(MEMORY_FILE, "r") as f:
        history = json.load(f)
else:
    history = []
```

```python
print("🎤 AI Interview Trainer")

print("Choose Role:")
print("1. AI/ML Engineer")
print("2. Data Analyst")
print("3. Python Developer")

choice = input("Enter choice: ")

if choice == "1":
    role = "AI/ML Engineer"

elif choice == "2":
    role = "Data Analyst"

else:
    role = "Python Developer"
```

```python
if len(history) > 0:

    print("\n📊 Previous Interview Records")

    for record in history:
        print(
            f"Date: {record['date']} | "
            f"Role: {record['role']} | "
            f"Score: {record['score']}"
        )

else:
    print("No previous interview records found.")
```

```python
if choice == "1":

    questions = [
        "What is Machine Learning?",
        "What is supervised learning?",
        "Difference between AI and ML?"
    ]

elif choice == "2":

    questions = [
        "What is data analysis?",
        "What is Pandas?",
        "Why is data visualization important?"
    ]

else:

    questions = [
        "What is Python?",
        "What are lists?",
        "What is a function?"
    ]
```

```python
answers = []

for question in questions:
    answer = input(question + "\n")
    answers.append(answer)
```

```python
score = 0

for answer in answers:

    answer = answer.strip()

    if len(answer.split()) >= 5:
        score += 1

print("Score:", score, "/", len(questions))
```

```python
if score == len(questions):

    print("🌟 Excellent Performance")

elif score >= 2:

    print("👍 Good Performance")

else:

    print("📚 More Practice Needed")
```

```python
new_record = {
    "date": datetime.now().strftime("%Y-%m-%d %H:%M"),
    "role": role,
    "score": score
}

history.append(new_record)

with open(MEMORY_FILE, "w") as f:
    json.dump(history, f, indent=4)

print("\n✅ Interview saved successfully!")
```

```python
print("Total Records:", len(history))

if len(history) >= 2:

    previous_score = history[-2]["score"]
    current_score = history[-1]["score"]

    improvement = current_score - previous_score

    print("\n📈 Progress Report")
    print("Previous Score:", previous_score)
    print("Current Score:", current_score)

    if improvement > 0:
        print("Improved by", improvement, "points 🎉")

    elif improvement < 0:
        print("Score decreased by", abs(improvement), "points")

    else:
        print("No change in score")

else:
    print("Take at least 2 interviews to see progress.")
```
